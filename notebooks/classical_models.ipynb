# CLASSICAL MODELS

```python
import numpy as np

X_train = np.load("../dataset/X_train.npy")
y_train = np.load("../dataset/y_train.npy")

X_val = np.load("../dataset/X_val.npy")
y_val = np.load("../dataset/y_val.npy")

X_test = np.load("../dataset/X_test.npy")
y_test = np.load("../dataset/y_test.npy")
```

```python
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_val_flat = X_val.reshape(X_val.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

print(X_train_flat.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_flat)
X_val_scaled = scaler.transform(X_val_flat)
X_test_scaled = scaler.transform(X_test_flat)
```

## LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model_lr = LogisticRegression(max_iter=1000, solver='lbfgs')

print("Training Logistic Regression...")
model_lr.fit(X_train_scaled, y_train)

# Validation
y_val_pred_lr = model_lr.predict(X_val_scaled)
val_acc_lr = accuracy_score(y_val, y_val_pred_lr)

# Test
y_test_pred_lr = model_lr.predict(X_test_scaled)
test_acc_lr = accuracy_score(y_test, y_test_pred_lr)

print("Logistic Regression Val Accuracy:", val_acc_lr)
print("Logistic Regression Test Accuracy:", test_acc_lr)
```

## LINEAR SVM

```python
from sklearn.svm import SVC

model_svm_linear = SVC(kernel='linear')

print("Training Linear SVM...")
model_svm_linear.fit(X_train_scaled, y_train)

# Validation
y_val_pred_svm_lin = model_svm_linear.predict(X_val_scaled)
val_acc_svm_lin = accuracy_score(y_val, y_val_pred_svm_lin)

# Test
y_test_pred_svm_lin = model_svm_linear.predict(X_test_scaled)
test_acc_svm_lin = accuracy_score(y_test, y_test_pred_svm_lin)

print("SVM Linear Val Accuracy:", val_acc_svm_lin)
print("SVM Linear Test Accuracy:", test_acc_svm_lin)
```

## PCA

```python
from sklearn.decomposition import PCA

PCA(n_components=100, random_state=42)

# ONLY fit on training data
X_train_pca = pca.fit_transform(X_train_scaled)

# ONLY transform others
X_val_pca = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

## RBF SVM

```python
model_svm_rbf = SVC(kernel='rbf', C=0.5, gamma='scale')

print("Training RBF SVM...")
model_svm_rbf.fit(X_train_pca, y_train)

# Validation
y_val_pred_svm_rbf = model_svm_rbf.predict(X_val_pca)
val_acc_svm_rbf = accuracy_score(y_val, y_val_pred_svm_rbf)

# Test
y_test_pred_svm_rbf = model_svm_rbf.predict(X_test_pca)
test_acc_svm_rbf = accuracy_score(y_test, y_test_pred_svm_rbf)

print("SVM RBF Val Accuracy:", val_acc_svm_rbf)
print("SVM RBF Test Accuracy:", test_acc_svm_rbf)
```

## RESULTS

```python
print("\n===== FINAL RESULTS =====")

print(f"Logistic Regression  | Val: {val_acc_lr:.4f} | Test: {test_acc_lr:.4f}")
print(f"SVM Linear           | Val: {val_acc_svm_lin:.4f} | Test: {test_acc_svm_lin:.4f}")
print(f"SVM RBF (PCA)        | Val: {val_acc_svm_rbf:.4f} | Test: {test_acc_svm_rbf:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['LogReg', 'SVM Linear', 'SVM RBF']
val_acc = [val_acc_lr, val_acc_svm_lin, val_acc_svm_rbf]
test_acc = [test_acc_lr, test_acc_svm_lin, test_acc_svm_rbf]

x = np.arange(len(models))
width = 0.35

plt.figure()
plt.bar(x - width/2, val_acc, width, label='Validation')
plt.bar(x + width/2, test_acc, width, label='Test')

plt.xticks(x, models)
plt.ylabel('Accuracy')
plt.title('Validation vs Test Accuracy (Generalisation)')
plt.legend()

plt.show()
```
